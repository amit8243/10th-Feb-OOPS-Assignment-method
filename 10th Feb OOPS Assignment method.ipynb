```python
#OOPS Assign method
"""Q1. Which function is used to open a file? What are the different modes of opening a file? Explain each mode
of file opening."""

Ans: The open() function is used to open a file in Python.

The different modes of opening a file in Python are: 

1. 'r' - This is the default mode. It opens the file for reading only. The file pointer is placed at the beginning of the  file.

2. 'w' - This mode opens the file for writing only. If the file already exists, its contents are overwritten. If the file does not exist, a new one is created.

3. 'a' - This mode opens the file for appending  only. If the file already exists, the data is appended to the end of the file. If the file does not exist, a new one is created. 


4. 'r+' - This mode opens the file for both reading and writing. The file pointer is placed at the beginning of the file.

5. 'w+' - This mode opens the file for both reading and writing. If the file already exists, its contents are overwritten. If the file does not exist, a new one is created.

6. 'a+' - This mode opens the file for both reading and writing. If the file already exists, the data is appended to the end of the file. If the file does not exist  a new one is created.

"""Q2. Why close() function is used? Why is it important to close a file?"""

Ans: The close() function is used in Python to close an open file. It is important to close a file 
because it frees up system resources associated with the file, and it ensures that any data written to 
the file is properly saved. Additionally, closing a file allows other programs to access the same  file. 

If a file is not closed properly, it can lead to data corruption and other issues. Therefore, it is 
important to always close a file after you are done using it. 

In Python, the close() function is used to close an open file. It is important to remember to call this 
function when you are done using a file, as it will help ensure that your data is properly saved and 
that other programs can access the same file.

"""Q3. Write a python program to create a text file. Write ‘I want to become a Data Scientist’ in that file. Then
close the file. Open this file and read the content of the file."""

Ans:  
# Open a file in write mode 
file = open("amit.txt", "w") 
  
# Write content in the file 
file.write("I want to become a Data Scientist")
  
# Close the file 
file.close() 
  
# Open a file in read mode 
file = open("amit.txt", "r") 
  
# Read content of the file 
content = file.read() 
  
# Print
print(content) 
  
# Close the file 
file.close()

"""Q4. Explain the following with python code: read(), readline() and readlines()."""

Ans: 
#Code Implementation with read()
f = open("demofile.txt", "r")
print(f.read())

#Code Implementation with readline()
#open the file for read operation
fl = open('pythonfile.txt')
# reads line by line
ln = fl.readline()
while ln!= "":
print(ln)
ln = fl.readline()
#close the file
fl.close()

#Code Implementation with readlines()
#open the file for read operation
fl = open('pythonfile.txt')
# reads line by line and stores them in list
for ln in fl.readlines():
print(ln)
#close the file
fl.close()

"""Q5. Explain why with statement is used with open(). What is the advantage of using with statement and
open() together?"""
Ans: The with statement is used with open() in Python to ensure that the file is properly closed after the block of code
within the with statement is executed, even if an exception is raised. This helps to prevent data loss and resource 
leaks. The advantage of using with statement and open() together in Python is that it provides a convenient way to 
ensure that the file is properly closed after the block of code within the with statement is executed, even if an 
exception is raised. This helps to prevent data loss and resource leaks.

Additionally, the with statement simplifies the code by encapsulating common tasks such as opening and closing files.
This makes the code more readable and easier to maintain.

"""Q6. Explain the write() and writelines() functions. Give a suitable example.""""
Ans: The write() function is used to write a string to an open file. It takes a single parameter, which is the string to
be written. The writelines() function is used to write a sequence of strings to an open file. It takes a single 
parameter, which is the sequence of strings to be written.

Example:

# Open a file in write mode 
f = open("test.txt", "w") 
  
# Writing to the file 
f.write("This is a test file\n") 
  
# List of strings 
L = ["This is Delhi \n", "This is Chennai \n", "This is Bangalore \n"] 
  
# Writing multiple lines 
f.writelines(L) 
  
# Closing the file 
f.close()
```
